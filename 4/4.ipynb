```python
def main():
    command = ''
    commands = {'p':p, 's':s, 'l':l, 'ads':ads, 'ds':ds, 'ad':ad, 'd':d, 'm':m}
    while True:
        command = input('Введите команду: ')
        if command in commands:
            commands[command]()
        elif command == 'q':
            break
        else:
            print('Команда не найдена')
        print('---------------------')
        
def p():
    inputDocNum = input('Введите номер документа: ')
    owner = findOwnerByDocNum(inputDocNum)
    if owner != None:
        print('Владелец документа:', owner)
    else:
        print('Документ не найден в базе') 
    
def findOwnerByDocNum(docNum):
    global documents
    
    for document in documents:
        if document['number'] == docNum:
            return document['name']
    return None

def s(inputDocNum = False):
    if inputDocNum == False:
        inputDocNum = input('Введите номер документа: ')
    
    dirNum = findDirNumByDocNum(inputDocNum)
    if dirNum:
        print('Документ хранится на полке:', dirNum)
    else:
        print('Документ не найден в базе')
    
def findDirNumByDocNum(docNum):
    global directories
    
    for dirNum in directories:
        for documentNum in directories[dirNum]:
            if documentNum == docNum:
                return dirNum
    return False
    
def l(needPrefix=False):
    global documents
    
    if needPrefix:
        print('Текущий список документов:')
    for document in documents:
        print('№:', document['number']+',', 'тип:', document['type']+',', 'владелец:', document['name']+',', 'полка хранения:', findDirNumByDocNum(document['number']))
        
def ads():
    global directories
    
    inputDir = input('Введите номер полки: ')
    
    if inputDir in directories:
        print('Такая полка уже существует.', directoriesNow(directories))
    else:
        directories[inputDir] = []
        print('Полка добавлена.', directoriesNow(directories) )

def directoriesNow(directories):
    return 'Текущий перечень полок: '+', '.join(list(directories.keys()))

def ds():
    global directories
    
    inputDir = input('Введите номер полки: ')
    
    if inputDir not in directories:
        print('Такой полки не существует.', directoriesNow(directories))
    elif directories[inputDir] != [] :
        print('На полке есть документы, удалите их перед удалением полки.', directoriesNow(directories))
    else:
        directories.pop(inputDir)
        print('Полка удалена.', directoriesNow(directories) )

def ad():
    global directories
    global documents
    
    inputDocNum = input('Введите номер документа: ')
    inputType = input('Введите тип документа: ')
    inputOwner = input('Введите владельца документа: ')
    inputDir = input('Введите полку для хранения: ')
    
    if inputDir not in directories:
        print('Такой полки не существует. Добавьте полку командой as.')
    else:
        documents.append({'type':inputType, 'number':inputDocNum, 'name': inputOwner})
        directories[inputDir].append(inputDocNum)
        print('Документ добавлен.')
    l(True)
    
def d():
    global directories
    global documents
    
    inputDocNum = input('Введите номер документа:')
    
    for key, document in enumerate(documents):
        if document['number'] == inputDocNum:
            documents.pop(key)
    if deleteDocFromDir(inputDocNum):
        print('Документ удален.')
    else:
        print('Документ не найден в базе.')
    l(True)

    
def deleteDocFromDir(docNum):
    directory = findDirNumByDocNum(docNum)
    if directory:
        for i, val in enumerate(directories[directory]):
            if directories[directory][i] == docNum:
                del directories[directory][i]
                return True
    return False
        
def m():
    global directories
    
    inputDocNum = input('Введите номер документа:')
    inputDir = input('Введите номер полки:')
    
    if inputDir not in directories:
        print('Такой полки не существует.', directoriesNow(directories))
        return
    
    if deleteDocFromDir(inputDocNum):
        directories[inputDir].append(inputDocNum)
        print('Документ перемещен.')
    else:
        print('Документ не найден в базе.')
    l(True)
        
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

main()
```
